```python
import os
import time

def get_page(language, url, index):
    try:
        driver.get(url)
    except:
        print(f"ERROR: Could not load {url}")
        return

    count = 0
    while True:
        if driver.execute_script("return document.readyState") == "complete":
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)
            break
        else:
            print("Waiting for the page to load...")
            time.sleep(10)
            count += 1
            if count >= 3:
                print(f"WARNING: document.readyState==false")
                break

    driver.execute_script("window.scrollBy(0, 1000);")

    html_source = driver.page_source

    dirname = f"websites/{language}"
    os.makedirs(dirname, exist_ok=True)

    # Save the HTML to a file
    html_file_path = f"{dirname}/{index}.html"  # Replace with your desired file path
    with open(html_file_path, "w", encoding="utf-8") as file:
        file.write(html_source)
    
    print(f"Saved URL: {url}")
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import os

# Function to recursively get all text nodes
def get_text_nodes(element):
    text_nodes = []
    if element.name is None:
        text_nodes.append(element.strip())
    elif element.name != 'script':
        for content in element.contents:
            if isinstance(content, str):
                text_nodes.append(content.strip())
            else:
                text_nodes.extend(get_text_nodes(content))
    return text_nodes
```

```python
datadir = "websites"
languages = { name: [] for name in os.listdir(datadir) }
for name, websites in languages.items():
    for path in os.listdir(f"{datadir}/{name}"):
        index = int(path.split(".")[0])

        with open(f"{datadir}/{name}/{path}", 'r') as file:
            data = file.read().rstrip()

        soup = BeautifulSoup(data, 'html.parser')
        body = soup.find('body')

        text_nodes = get_text_nodes(body)

        while index >= len(websites):
            websites.append([])

        websites[index] = text_nodes

    
```

```python
from enum import Enum

class StringType(Enum):
    ASCII = 1
    LATIN1 = 2
    UTF16 = 3

def check_string(s: str):
    is_ascii = True

    if len(s) == 0:
        return StringType.ASCII

    for c in s:
        # print(ord(c), c)

        c = ord(c)
        if c > 256:
            return StringType.UTF16
        elif c > 128:
            is_ascii = False
    
    if is_ascii:
        return StringType.ASCII
    else:
        return StringType.LATIN1
```

```python
for name, websites in languages.items():
    print(f"Processing language {name}...")
    for website in websites:
        for i in range(len(website)):
            website[i] = check_string(website[i])
```

```python
for name, websites in languages.items():
    print(f"Processing language {name}...")
    for i, website in enumerate(websites):
        if len(website) == 0:
            websites[i] = None
            continue

        ascii_count = 0
        latin1_count = 0
        utf16_count = 0
        for string in website:
            if string == StringType.ASCII:
                ascii_count += 1
            elif string == StringType.LATIN1:
                latin1_count += 1
            elif string == StringType.UTF16:
                utf16_count += 1
            else:
                assert False, "Unknown string type"

        websites[i] = {
            'ascii': ascii_count,
            'latin1': latin1_count,
            'utf16': utf16_count,
        }
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame()
for name, websites in languages.items():
    df = pd.concat([df, pd.DataFrame({
        f"{name.lower()}_ascii": np.empty(len(websites)),
        f"{name.lower()}_latin1": np.empty(len(websites)),
        f"{name.lower()}_uft16": np.empty(len(websites))
    })])

for name, websites in languages.items():
    for i, website in enumerate(websites):
        df.loc[i, f"{name.lower()}_ascii"] = website["ascii"] if website else float("nan")
        df.loc[i, f"{name.lower()}_latin1"] = website["latin1"] if website else float("nan")
        df.loc[i, f"{name.lower()}_uft16"] = website["utf16"] if website else float("nan")

df.to_csv("data.csv")
df.head(5)
```

```python
df = df.apply(lambda col: col.fillna(int(col.mean())), axis=0)
```

```python
new_df = pd.DataFrame(columns={ name: [0] for name in df.columns })
for name in df.columns:
    new_df.loc[0, name] = np.sum(df[name])
```

```python
percent = { }
for name in languages.keys():
    ascii_count = new_df[f"{name.lower()}_ascii"]
    latin1_count = new_df[f"{name.lower()}_latin1"]
    utf16_count = new_df[f"{name.lower()}_uft16"]
    percent[f"{name} Ascii"] = (ascii_count) / (ascii_count + latin1_count + utf16_count)
    percent[f"{name} Latin1"] = (ascii_count + latin1_count) / (ascii_count + latin1_count + utf16_count)
    percent[f"{name} Non-Ascii Latin1"] = latin1_count / (ascii_count + latin1_count + utf16_count)
    percent[f"{name} UTF16"] = (utf16_count) / (ascii_count + latin1_count + utf16_count)

df = pd.DataFrame(percent)
```
